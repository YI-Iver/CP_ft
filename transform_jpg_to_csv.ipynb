```python
import cv2
import numpy as np
import pandas as pd
import os
from sklearn import utils
from tqdm.notebook import tqdm
```

```python
train_dir = r'C:\Users\Iver_Home\CP_ft\mix_test'
```

```python
def labeling(img):
    class_name = img.split('.')[0]
    if class_name == 'mevius':
        return 1
    elif class_name =='marlboro':
        return 2
    elif class_name =='esse':
        return 3
    elif class_name =='bohem':
        return 4
    elif class_name =='dunhill':
        return 5
    
```

```python
print(labeling('esse.14.jpg'))
```

```python
x_data = list()
t_data = list()
```

```python
print(os.listdir(train_dir))
arr127 = np.full((252,189),127)

```

```python

for img in tqdm(os.listdir(train_dir)):
    total = len(os.listdir(train_dir))   #=100
    position = 0
    leave = True
    label_data = labeling(img)
    img_path = os.path.join(train_dir,img)
    img_data = cv2.resize(cv2.imread(img_path,cv2.IMREAD_GRAYSCALE),(189,252))
    #img_data : 2d array
    img_data = img_data + arr127
    x_data.append(img_data.ravel())
    t_data.append(label_data)
print(img_data)
```

```python
x_df = pd.DataFrame(x_data)
t_df = pd.DataFrame({'label':t_data})
display(x_df)
```

```python
df = pd.merge(t_df, x_df, left_index=True, right_index=True)
display(x_df)
```

```python
shuffled_df = utils.shuffle(df)
```

```python
shuffled_df.to_csv(r'C:\Users\Iver_Home\CP_ft\mix_output\test_15.csv', index=False)
#df.to_csv(r'C:\Users\Iver_Home\CP_ft\mix_output\test_all2.csv', index=False)
```
